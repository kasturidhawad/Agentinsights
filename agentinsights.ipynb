```python
import pandas as pd
from transformers import AutoTokenizer, T5ForConditionalGeneration
```

```python
data = pd.read_excel('Summ.xlsx')
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("t5-small")
```

```python
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")
def generate_summary(article):
    # Add a space after "summarize:" in the input
    input_text = "summarize: " + article
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(input_ids, max_length=150, num_beams=4, length_penalty=2.0, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

data['t5'] = data['article'].apply(generate_summary)
```

```python
output_file_path = 'Summ.xlsx'  # Change to your desired output file name
data.to_excel(output_file_path, index=False)

print(f'Summaries saved')
```

```python
import pandas as pd
from transformers import BertTokenizer, BertForMaskedLM
import torch
```

```python
# Initialize BERT Model
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForMaskedLM.from_pretrained(model_name)

# Generate Summaries
summaries = []

for text in data['article']:
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs, return_dict=True)
    summary = tokenizer.decode(torch.argmax(outputs.logits, dim=2)[0], skip_special_tokens=True)
    summaries.append(summary)

# Add the generated summaries to the DataFrame
data['BERT'] = summaries

# Save the updated data to an Excel sheet
data.to_excel('Summ.xlsx', index=False)
```

```python
import pandas as pd
from rouge_score import rouge_scorer
data = pd.read_excel('Summ.xlsx')


scorer = rouge_scorer.RougeScorer(['rougeLsum'])

# Initialize a list to store the Rouge-Lsum scores
rougeLsum_scores = []


for index, row in data.iterrows():

    highlights = row['highlights']
    t5 = row['t5']

    # Calculate Rouge-Lsum scores
    scores = scorer.score(highlights, t5)

    rougeLsum_scores.append(scores['rougeLsum'].fmeasure)

data['rougeLsum_t5'] = rougeLsum_scores

data.to_excel('Summ.xlsx', index=False)
```

```python
import pandas as pd
from rouge_score import rouge_scorer

data = pd.read_excel('Summ.xlsx')
scorer = rouge_scorer.RougeScorer(['rougeLsum'])

rougeLsum_scores_BART = []
rougeLsum_scores_BERT = []

for index, row in data.iterrows():
    highlights = row['highlights']

    BART = row['BART']
    BERT = row['BERT']

    # Calculate Rouge-Lsum scores for 'BART'
    scores_BART = scorer.score(highlights, BART)
    rougeLsum_scores_BART.append(scores_BART['rougeLsum'].fmeasure)

    # Calculate Rouge-Lsum scores for 'BERT'
    scores_BERT = scorer.score(highlights, BERT)
    rougeLsum_scores_BERT.append(scores_BERT['rougeLsum'].fmeasure)

data['rougeLsum_BART'] = rougeLsum_scores_BART
data['rougeLsum_BERT'] = rougeLsum_scores_BERT

# Save the updated DataFrame with Rouge-Lsum scores to the Excel file
data.to_excel('Summ.xlsx', index=False)
```
